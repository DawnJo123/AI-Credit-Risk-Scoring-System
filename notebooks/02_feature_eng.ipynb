```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("../data/cleaned_data/cleaned_data.csv")
df.shape
```

```python
df['fico_avg'] =(df['fico_range_low']+df['fico_range_high'])/2
```

```python
df['loan_to_income']=df['loan_amnt']/df['annual_inc']
```

```python
df["installment_to_income"]=(df['installment']*12)/ df['annual_inc']
```

```python
df['credit_per_account']=df['revol_bal'] / df['open_acc']
```
